```python
import base64

# 이미지 파일 열기 (바이너리 모드)
with open("resource\KakaoTalk_20250711_160050558.jpg", "rb") as image_file:
    image_bytes = image_file.read()

# Base64로 인코딩
image_base64 = base64.b64encode(image_bytes).decode("utf-8")

# 텍스트 파일에 저장
with open("Photo_num/myD.txt", "w", encoding="utf-8") as f:
    f.write(image_base64)
```
